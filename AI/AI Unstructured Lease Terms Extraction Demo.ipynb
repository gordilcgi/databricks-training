```sql
SELECT
ai_parse_document(
  content,
  map('version', '2.0')
) AS parsed
FROM READ_FILES('/Volumes/workspace/default/testdata/leases/', format => 'binaryFile');
```

```sql
WITH leases AS (
  SELECT
    path,
    ai_parse_document(content) AS parsed
  FROM
    READ_FILES('/Volumes/workspace/default/testdata/leases/', format => 'binaryFile')
)
SELECT
  path,
  ai_extract(
    parsed,
    '{
      "start_date": {"type": "string", "description": "Lease start date in MM/DD/YYYY format"},
      "end_date": {"type": "string", "description": "Lease end date in MM/DD/YYYY format"},
      "lease_amount": {"type": "number", "description": "Monthly rent amount in dollars"}
    }',
    MAP('instructions', 'Extract key lease terms from this rental agreement.')
  ) AS lease_terms
FROM leases;
```
